Lets do Label Encoding now

```python
# Smoking                       2
# Family History of Anxiety     2
# Dizziness                     2
# Medication                    2
# Recent Major Life Event       2
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()

columns_to_encode = ['Smoking', 'Family History of Anxiety', 'Dizziness', 'Medication', 'Recent Major Life Event']

for column in columns_to_encode:
    newdf[column] = labelEncoder.fit_transform(newdf[column])

print(newdf.head())
```

```python
x,y=newdf.drop(columns=['Occupation','Gender','Anxiety Level (1-10)'],axis=1), newdf['Anxiety Level (1-10)']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score

rand_for = RandomForestRegressor(random_state = 99).fit(x_train, y_train)
```
